```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.patches as mpatches
labels = ['Red Bar', 'Magenta Bar', 'Cyan Bar', 'Black Bar']
```

```python
AMZN['Adj Close'].plot(label='AMZN',figsize=(16,8),title='Adjusted Closing Price')
MSFT['Adj Close'].plot(label='MSFT')
GOOGL['Adj Close'].plot(label='GOOGL')

plt.title('Adjusted Closing Price')
plt.xlabel('Timeline') #this is the label name
plt.ylabel('Price ($)') #this is the label name

patch1 = mpatches.Patch(color='orange', label='Microsoft')
patch2 = mpatches.Patch(color='blue', label='Amazon')
patch3 = mpatches.Patch(color='Green', label='Google')

plt.legend(handles=[patch1, patch2, patch3])

plt.show()
```

```python
AMZN['MA50'] = AMZN['Adj Close'].rolling(50).mean()
AMZN['MA200'] = AMZN['Adj Close'].rolling(200).mean()
AMZN[['Adj Close','MA50','MA200']].plot(label='AMAZON',figsize=(16,8))

plt.title('Graph Showing Rolling Price Over Time for Amazon')
plt.xlabel('Timeline') #this is the label name
plt.ylabel('Price ($)') #this is the label name
```

```python
MSFT['MA50'] = MSFT['Adj Close'].rolling(50).mean()
MSFT['MA200'] = MSFT['Adj Close'].rolling(200).mean()
MSFT[['Adj Close','MA50','MA200']].plot(label='MICROSOFT',figsize=(16,8))

plt.title('Graph Showing Rolling Price Over Time for Microsoft')
plt.xlabel('Timeline') #this is the label name
plt.ylabel('Price ($)') #this is the label name
```

```python
GOOGL['MA50'] = GOOGL['Adj Close'].rolling(50).mean()
GOOGL['MA200'] = GOOGL['Adj Close'].rolling(200).mean()
GOOGL[['Adj Close','MA50','MA200']].plot(label='GOOGLE',figsize=(16,8))

plt.title('Graph Showing Rolling Price Over Time for Google')
plt.xlabel('Timeline') #this is the label name
plt.ylabel('Price ($)') #this is the label name
```
