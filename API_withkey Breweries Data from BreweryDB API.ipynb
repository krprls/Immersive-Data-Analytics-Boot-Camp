```python
#create empty lists to extract data

idls = [] #ID of the beer
namels = [] #name of the beer
establishedls = []# The year established
statusls = [] # status of the breweries product
isInBusinessls = []#Breweries in business
isMassOwnedls = [] # Breveries is produced in mass
organicls = [] #Y/N if it is organic
websitels = [] #product detials in website
descriptionls = [] # description of brewerie
```

```python
    #extract the values for the bbreweries dataset

    #get the value of each key and if the key doesn't exist, set a variable to be None

for brewery in breweries:
        try: bid = brewery['id']
        except: bid = None
        
        try: name = brewery['name']
        except: name = None
            
        try: established = brewery['established']
        except: established = None
            
        try: status = brewery['status']
        except: status = None
            
        try: Business = brewery['isInBusiness']
        except: Business = None
            
        try: MassOwned = brewery['isMassOwned']
        except:MassOwned = None
            
        try: organic = brewery['isOrganic']
        except:  organic = None
    
        try: website = brewery['website']
        except: website = None
                            
        try: descr = brewery['description']
        except: descr = None
        
        idls.append(bid)
        namels.append(name)
        establishedls.append(established)
        statusls.append(status)
        isInBusinessls.append(Business)
        isMassOwnedls.append(MassOwned )
        organicls.append(organic)
        websitels.append(website)                        
        descriptionls.append(descr)
   
```

```python
#match indices of each list and zip into one list
breweryList = list(zip(idls, namels, establishedls, 
                    statusls, isInBusinessls, isMassOwnedls,
                     organicls, websitels, descriptionls))

#names for columns in data frame
colnames = ['id', 'name', 'establishedls','statusls', 'isInBusinessls', 'isMassOwnedls','organic',  'websitels', 'description'
            ]
```

```python
#create data frame with column names in it
df = pd.DataFrame(breweryList , columns=colnames)

df.head(10)
```
